```python
# all of the imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso, LassoCV
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
# Read in Data

movs_df = pd.read_csv("data/processed_movie_database.csv")
movs_df.drop(columns="Unnamed: 0", inplace=True)
```

```python
movs_df.head()
```

```python
# Pull in features and label

X_smooth = movs_df[["Production Budget", "Runtime", "Release Year",
                    "Actor Trendy Score", "Director Trendy Score",
                    "Producer Trendy Score", "Genre Trendy Score"]]
X_dummies = movs_df["Release Month"].str.get_dummies()
X_dummies2 = movs_df["MPAA Rating"].str.get_dummies()
X = pd.concat([X_smooth, X_dummies, X_dummies2], axis=1)
y = movs_df["Domestic Total Gross"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Doing 5 kfolds for cross-validation
kfold = KFold(n_splits=5, shuffle=True)
```

```python
sns.pairplot(pd.concat([X_smooth, y], axis=1));
```

```python
sns.distplot(y_train)
```

### Thoughts:

1. As we can see the distribution for the Domestic Total Gross is very tail-heavy, indicating the presence of outliers.  This is perfectly expected in the movie business, since most highly successful movies are outliers.  It's worth remembering here that the movie business exists because even though most movies well-underperform, enough movies way, way overperform that money is made on average.

2. The Production Budget, Runtime, and Actor Score look like they could be improved with quadratic terms (probably very low coefficients, but they might make modest improvements) since they seem to tail upwards very slightly.  Quadratic terms will likely modestly improve the model.

---
# Basic Test Models: Linear Regression, Ridge, Lasso

```python
# See how a baseline LinearRegression performs

lm = LinearRegression()

lm.fit(X_train, y_train)
scores = cross_val_score(lm,
                         X_test[["Production Budget", "Runtime", "Release Year",
                                "Actor Trendy Score", "Director Trendy Score",
                                "Producer Trendy Score", "Genre Trendy Score"]],
                         y_test, cv=kfold)
print(scores)

avg_score = sum(scores) / len(scores)
avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

```python
# See how a Ridge alpha=10.0 performs

ridge = Ridge(alpha=10.0)

ridge.fit(X_train, y_train)
scores = cross_val_score(ridge, X_train, y_train, cv=kfold, scoring="r2")
print(scores)

avg_score = sum(scores) / len(scores)
avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

```python
# See how a Ridge alpha=1.0 performs

ridge = Ridge(alpha=1.0)

ridge.fit(X_train, y_train)
scores = cross_val_score(ridge, X_train, y_train, cv=kfold)
print(scores)

avg_score = sum(scores)/len(scores)

avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

```python
# See how a Lasso at alpha=10.0 performs

lasso = Lasso(alpha=10.0)

lasso.fit(X_train, y_train)
scores = cross_val_score(lasso, X_train, y_train, cv=kfold)
print(scores)

avg_score = sum(scores)/len(scores)

avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

```python
# See how a Lasso at alpha=1.0 performs

lasso = Lasso(alpha=1.0)

lasso.fit(X_train, y_train)
scores = cross_val_score(lasso, X_test, y_test, cv=kfold)
print(scores)

avg_score = sum(scores)/len(scores)

avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

```python
# See how a Lasso at alpha=0.1 performs

lasso = Lasso(alpha=0.1)

lasso.fit(X_train, y_train)
scores = cross_val_score(lasso, X_train, y_train, cv=kfold)
print(scores)

avg_score = sum(scores)/len(scores)

avg_score, (sum((scores - avg_score)**2) / len(scores))**0.5
```

# Residuals for Best Basic Model

Here I run a residual plot for the best basic model, the Ridge for alpha=10.0.

```python
ridge = RidgeCV(cv=kfold, normalize=True)
ridge.fit(X_train, y_train)


y_pred = ridge.predict(X_train)

plt.scatter(y_pred, y_train-y_pred, alpha=.4)
plt.plot(np.linspace(0,425,1000),np.linspace(0,0,1000))
plt.title('Residuals of the Model for Movie Domestic Total Gross')
plt.xlabel('Predicted Value')
plt.ylabel('Residual');
```
